```python
import numpy as np
import pandas as pd

df = pd.read_csv('task3df.csv')
```

```python
from datetime import datetime
dates = {
    "Mar 1": datetime(2020, 3, 1).timetuple().tm_yday,
    "May 31": datetime(2020, 5, 31).timetuple().tm_yday,
    "Jun 1": datetime(2020, 6, 1).timetuple().tm_yday,
    "Aug 31": datetime(2020, 8, 31).timetuple().tm_yday,
    "Sep 1": datetime(2020, 9, 1).timetuple().tm_yday,
    "Nov 30": datetime(2020, 11, 30).timetuple().tm_yday,
    "Dec 1": datetime(2020, 12, 1).timetuple().tm_yday,
    "Feb 29": datetime(2020, 2, 29).timetuple().tm_yday,
}

for key in dates.keys():
    print(f"{key}: {dates[key]}")
```

```python
df['SEASON'] = np.NaN
d = dates

df.loc[(d["Mar 1"] <= df['DISCOVERY_DOY']) & (df['DISCOVERY_DOY'] <= d["May 31"]), 'SEASON'] = "spring"
df.loc[(d["Jun 1"] <= df['DISCOVERY_DOY']) & (df['DISCOVERY_DOY'] <= d["Aug 31"]), 'SEASON'] = "summer"
df.loc[(d["Sep 1"] <= df['DISCOVERY_DOY']) & (df['DISCOVERY_DOY'] <= d["Nov 30"]), 'SEASON'] = "fall"
df.loc[(d["Dec 1"] <= df['DISCOVERY_DOY']) | (df['DISCOVERY_DOY'] <= d["Feb 29"]), 'SEASON'] = "winter"
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df['SEASON'])
df = df.drop(columns=['SEASON'])
df = df.join(dummies)
df.head()
```

```python
df.to_csv('task3df_v.csv')
```
